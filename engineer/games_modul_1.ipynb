### Общее распределение игр по жанрам. Что можно сказать о самых прибыльных жанрах? Выделяются ли жанры с высокими и низкими продажами?

Посмотрим на распределение игр по жанрам и на общие продажи по жанрам за актуальный период (в датасете `data_current`).

````python
games_by_genre = (
    data_current
    .pivot_table(index='genre', values='total_sales', aggfunc=['sum', 'count'])
    .droplevel(1, axis=1)
    .sort_values(by='sum', ascending=False)
)
games_by_genre.head()
````

Посчитаем средние продажи по игре в жанре.

````python
games_by_genre['mean_sum_per_game'] = games_by_genre['sum'] / games_by_genre['count']
games_by_genre.head()
````

Переведем данные в удобный формат для построения графиков в цикле.

````python
games_by_genre_format = (
    pd.melt(games_by_genre,
            value_vars=['sum', 'count', 'mean_sum_per_game'], 
            ignore_index=False)
)
games_by_genre_format.head()
````

Построим круговые диаграммы для оценки долей.

````python
for name, group_data in games_by_genre_format.groupby('variable'):
    (
        group_data
        .plot(title=name, kind='pie', y='value', figsize=(7, 7), autopct='%1.0f%%')
    )
plt.suptitle('Рисунок 16 - Круговые диаграммы для оценки долей', y=0.1);
````

На круговых диаграммах можно заметить, что доля средниех продаж по игре в Action (диаграмма mean_sum_per_game) значительно меньше, чем доля суммы продаж Action (диаграмма sum). Это можно объяснить следующим: ранее мы анализировали общие продажи на аномалии, заметили редкие большие значения. Например, сумма продаж одной игры в Action составляет 82,5 млн, что сильно выделяется из распределения продаж.

Построим столбиковые диаграммы.

````python
for name, group_data in games_by_genre_format.groupby('variable'):
    (
        group_data
        .reset_index()
        .sort_values(by='value', ascending=False)
        .plot(title=name, kind='bar', x='genre', y='value')
    )
plt.suptitle('Рисунок 17 - Распределение игр по жанрам', y=-0.2);
````

## Проверка гипотез

Зададим самостоятельно пороговое значение alpha.

Проверим гипотезы на данных [data_current](#data_current) (это данные за актуальный период, который мы выделили на прошлом этапе).

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;

Сформулируем нулевую и альтернативную гипотезы и зададим пороговое значение alpha.
```
H_0: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые`
H_a: Средние пользовательские рейтинги платформ Xbox One и PC отличаются`
alpha = 0.05
```

Проверим размер выборок.

````python
sample_1 = (
    data_current
    .dropna(subset=['user_score'])
    .query('platform == "XOne"')['user_score']
)
len(sample_1)
````

````python
sample_2 = (
    data_current
    .dropna(subset=['user_score'])
    .query('platform == "PC"')['user_score']
)

len(sample_2)
````

Выборки разного размера — укажем `equal_var= False` при вызове метода `scipy.stats.ttest_ind()`

````python
# значение уровеня статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу
alpha = .05 

# results = вызов метода для проверки гипотезы
results = st.ttest_ind(
    sample_1,
    sample_2,
    equal_var=False)

# вывод значения p-value на экран 
print('p-значение:', results.pvalue)

# условный оператор с выводом строки с ответом
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
````

### Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

Сформулируем нулевую и альтернативную гипотезы и зададим пороговое значение alpha
```
H_0: Средние пользовательские рейтинги жанров Action и Sports одинаковые`
H_1: Средние пользовательские рейтинги жанров Action и Sports разные`
alpha = 0.05
```

Проверим размер выборок.

````python
sample_1 = (
    data_current
    .dropna(subset=['user_score'])
    .query('genre == "Action"')['user_score']
)
len(sample_1)
````

````python
sample_2 = (
    data_current
    .dropna(subset=['user_score'])
    .query('genre == "Sports"')['user_score']
)
len(sample_2)
````

Выборки разного размера — укажем `equal_var= False` при вызове метода `scipy.stats.ttest_ind()`

````python
# значение уровеня статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу
alpha = .05 

# results = вызов метода для проверки гипотезы
results = st.ttest_ind(
    sample_1,
    sample_2,
    equal_var=False)

# вывод значения p-value на экран 
print('p-значение:', results.pvalue)

# условный оператор с выводом строки с ответом
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
````
